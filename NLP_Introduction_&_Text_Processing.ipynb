## Question 1: What is Computational Linguistics and how does it relate to NLP?

**Computational Linguistics (CL)** is an interdisciplinary field that deals with the statistical and rule-based modeling of natural language from a computational perspective. It involves the use of computer science techniques to study and process human languages.

**Natural Language Processing (NLP)** is a subfield of Artificial Intelligence (AI) that focuses on enabling computers to understand, interpret, and generate human language in a valuable way. NLP aims to bridge the gap between human communication and computer understanding.

#### Relationship:

Computational Linguistics provides the theoretical and analytical foundations for NLP. While CL is more focused on the scientific understanding of language through computational methods, NLP is more concerned with the practical applications of these methods to build systems that can perform language-related tasks (e.g., machine translation, sentiment analysis, chatbots). In essence, CL offers the 'what' and 'how' of language computation, while NLP leverages that knowledge for engineering solutions.

## Question 2: Briefly describe the historical evolution of Natural Language Processing.

The historical evolution of Natural Language Processing can be broadly categorized into several eras:

*   **Symbolic/Rule-Based Era (1950s - 1980s):** Early NLP efforts were primarily driven by symbolic approaches, where linguists and computer scientists crafted explicit rules to parse sentences, identify grammatical structures, and translate languages. Key milestones include the Georgetown experiment of 1954 (early machine translation) and the development of transformational grammar by Chomsky.

*   **Statistical Era (Late 1980s - Early 2000s):** With the rise of computational power and the availability of larger text corpora, NLP shifted towards statistical methods. Techniques like Hidden Markov Models (HMMs), Support Vector Machines (SVMs), and Conditional Random Fields (CRFs) became prominent. This era saw improvements in tasks like part-of-speech tagging, machine translation, and information extraction, relying heavily on probability and statistics to learn patterns from data rather than handcrafted rules.

*   **Machine Learning Era (Early 2000s - 2010s):** This period saw the integration of more sophisticated machine learning algorithms. Feature engineering was a crucial aspect, where experts designed relevant features for algorithms to learn from. Algorithms like Maximum Entropy and early neural networks started gaining traction, further improving performance in various NLP tasks.

*   **Deep Learning Era (2010s - Present):** The advent of deep neural networks, particularly recurrent neural networks (RNNs), convolutional neural networks (CNNs), and later transformer architectures, revolutionized NLP. Deep learning models can automatically learn complex features from raw text data, eliminating the need for manual feature engineering. Breakthroughs in word embeddings (e.g., Word2Vec, GloVe), and later large pre-trained language models (e.g., BERT, GPT-3/4), have led to significant advancements in tasks such as sentiment analysis, question answering, summarization, and text generation, achieving state-of-the-art results and pushing the boundaries of what machines can do with human language.

## Question 4: What is text normalization and why is it essential in text processing tasks?

**Text normalization** is the process of transforming text into a canonical (standard) form. This involves a series of steps to reduce variations in the text, making it consistent and easier for machines to process. The goal is to ensure that words or phrases that mean the same thing are represented in the same way, regardless of how they originally appeared.

**Why is it essential in text processing tasks?**

Text normalization is crucial for several reasons in NLP and text processing:

1.  **Consistency:** It brings uniformity to text data. For example, without normalization, 'USA', 'U.S.A.', and 'United States' might be treated as distinct entities, even though they refer to the same country.

2.  **Improved Accuracy:** By standardizing text, it reduces the complexity and noise in the data, which can significantly improve the accuracy and performance of downstream NLP tasks such as information retrieval, machine translation, sentiment analysis, and text classification. Algorithms can learn more effectively from consistent data.

3.  **Reduced Vocabulary Size:** Normalization techniques like stemming and lemmatization reduce words to their root forms, which helps in decreasing the overall vocabulary size. A smaller vocabulary makes models more efficient and less prone to overfitting.

4.  **Effective Matching and Comparison:** It enables more effective matching and comparison of text. For instance, in search engines, if a user searches for 'running', documents containing 'run', 'ran', or 'runner' might also be relevant. Normalization helps in identifying these relationships.

5.  **Handling Irregularities:** It helps in handling common irregularities such as capitalization, punctuation, spelling variations, and numerical representations, which can otherwise confuse NLP models.

Common techniques involved in text normalization include:

*   **Lowercasing:** Converting all text to lowercase (e.g., 'Apple' and 'apple' become 'apple').
*   **Punctuation Removal/Standardization:** Removing or standardizing punctuation marks.
*   **Tokenization:** Breaking text into smaller units (words or subwords).
*   **Stop Word Removal:** Eliminating common words (e.g., 'the', 'a', 'is') that often carry little meaning.
*   **Stemming:** Reducing words to their root or base form (e.g., 'running', 'ran' → 'run').
*   **Lemmatization:** Reducing words to their dictionary form (lemma), considering context (e.g., 'better' → 'good').
*   **Spell Correction:** Correcting misspelled words.
*   **Number Standardization:** Converting numerical representations to a consistent format.

In summary, text normalization is a foundational step that prepares raw text data for more advanced analysis, ensuring that NLP models work with clean, consistent, and meaningful input.

## Question 5: Compare and contrast stemming and lemmatization with suitable examples.

**Stemming** and **Lemmatization** are both techniques used in Natural Language Processing (NLP) to reduce words to their base or root form. This process, often part of text normalization, helps in reducing vocabulary size and improving the accuracy of various NLP tasks by treating different inflected forms of a word as a single item. While their goal is similar, their approaches and outcomes differ significantly.

#### Stemming

*   **Definition:** Stemming is a crude heuristic process that chops off the ends of words in the hope of achieving the goal correctly most of the time. It often results in a 'stem' that is not a proper dictionary word.
*   **Approach:** It works by removing suffixes from words based on a set of rules without considering the context or grammatical structure of the word. Algorithms like Porter Stemmer and Snowball Stemmer are commonly used.
*   **Pros:** Simpler and faster to implement, computationally less expensive.
*   **Cons:** Can produce non-dictionary words (e.g., 'automat' from 'automatically'), often less accurate, especially for irregular words.
*   **Example:**
    *   `running` → `run`
    *   `ran` → `ran` (might not be stemmed correctly by simple stemmers)
    *   `studies` → `studi`
    *   `cats` → `cat`
    *   `beautiful` → `beauti`

#### Lemmatization

*   **Definition:** Lemmatization is a more sophisticated process that correctly determines the base or dictionary form of a word, known as the 'lemma'. It always produces a valid word.
*   **Approach:** It considers the word's morphological analysis to derive its base form, taking into account the word's part of speech and context. It often uses a vocabulary and morphological analysis of words.
*   **Pros:** More accurate, produces valid dictionary words, handles irregular forms better.
*   **Cons:** More complex and computationally intensive, requires a lexicon and knowledge of linguistic rules.
*   **Example:**
    *   `running` → `run`
    *   `ran` → `run`
    *   `studies` → `study`
    *   `better` → `good`
    *   `am`, `are`, `is` → `be`

#### Comparison and Contrast

| Feature           | Stemming                                | Lemmatization                                      |
| :---------------- | :-------------------------------------- | :------------------------------------------------- |
| **Accuracy**      | Lower (heuristic-based)                 | Higher (linguistic rule-based)                     |
| **Output**        | Can be a non-dictionary word (stem)     | Always a dictionary word (lemma)                   |
| **Speed**         | Faster                                  | Slower                                             |
| **Complexity**    | Simpler (rule-based suffix stripping)   | More complex (requires lexicon & morphological analysis) |
| **Context**       | Generally ignores word's context        | Considers word's context and part of speech        |
| **Use Case**      | Information retrieval, spam detection   | Text summarization, question answering, machine translation |

In summary, while both stemming and lemmatization aim to reduce words to their root forms, lemmatization provides a more accurate and linguistically informed result by considering the word's context and meaning, making it suitable for tasks requiring higher precision. Stemming, being simpler and faster, is often preferred when speed is critical and a perfect root form isn't strictly necessary.

##Question 6: Write a Python program that uses regular expressions (regex) to extract all email addresses from the following block of text:

#####“Hello team, please contact us at support@xyz.com for technical issues, or reach out to our HR at hr@xyz.com. You can also connect with John at john.doe@xyz.org and jenny via jenny_clarke126@mail.co.us. For partnership inquiries, email partners@xyz.biz.”

```python
import re

text_block = (
    "Hello team, please contact us at support@xyz.com for technical issues, "
    "or reach out to our HR at hr@xyz.com. You can also connect with John at "
    "john.doe@xyz.org and jenny via jenny_clarke126@mail.co.us. For partnership "
    "inquiries, email partners@xyz.biz."
)

# Regular expression pattern to find email addresses
# This pattern covers common email formats, including subdomains and numbers/underscores
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

# Find all email addresses in the text block
email_addresses = re.findall(email_pattern, text_block)

print("Extracted Email Addresses:")
for email in email_addresses:
    print(email)
```

##Question 7: Given the sample paragraph below, perform string tokenization and frequency distribution using Python and NLTK:
#####“Natural Language Processing (NLP) is a fascinating field that combines linguistics, computer science, and artificial intelligence. It enables machines to understand, interpret, and generate human language. Applications of NLP include chatbots, sentiment analysis, and machine translation. As technology advances, the role of NLP in modern solutions is becoming increasingly critical.”

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Download necessary NLTK data (only needs to be run once)
try:
    nltk.data.find('tokenizers/punkt_tab') # Check for punkt_tab specifically
except LookupError:
    nltk.download('punkt')
    nltk.download('punkt_tab') # Download punkt_tab if not found

paragraph = (
    "Natural Language Processing (NLP) is a fascinating field that combines "
    "linguistics, computer science, and artificial intelligence. It enables "
    "machines to understand, interpret, and generate human language. "
    "Applications of NLP include chatbots, sentiment analysis, and machine "
    "translation. As technology advances, the role of NLP in modern solutions "
    "is becoming increasingly critical."
)

# Step 1: String Tokenization
# Convert to lowercase to treat words like 'Natural' and 'natural' as the same
words = word_tokenize(paragraph.lower())

print("--- Tokenized Words ---")
print(words)
print("\n")

# Step 2: Frequency Distribution
# Filter out punctuation and non-alphabetic tokens for cleaner frequency distribution
filtered_words = [word for word in words if word.isalpha()]
freq_dist = FreqDist(filtered_words)

print("--- Frequency Distribution (Top 10) ---")
# Display the top 10 most common words
for word, frequency in freq_dist.most_common(10):
    print(f"{word}: {frequency}")
```

##Question 8: Create a custom annotator using spaCy or NLTK that identifies and labels proper nouns in a given text.

```python
import spacy

# Load the English language model
try:
    nlp = spacy.load('en_core_web_sm')
except OSError:
    print("SpaCy model 'en_core_web_sm' not found. Please run the download command: !python -m spacy download en_core_web_sm")
    exit()

def find_proper_nouns(text):
    """
    Identifies and labels proper nouns in a given text using spaCy.
    """
    doc = nlp(text)
    proper_nouns = []
    for token in doc:
        # 'PROPN' is the Part-of-Speech tag for proper nouns
        if token.pos_ == "PROPN":
            proper_nouns.append({"text": token.text, "label": token.pos_})
    return proper_nouns

# Sample text
sample_text = "Google was founded by Larry Page and Sergey Brin. It is headquartered in Mountain View, California."

# Identify proper nouns
identified_proper_nouns = find_proper_nouns(sample_text)

print("Identified Proper Nouns:")
for pn in identified_proper_nouns:
    print(f"  - Text: {pn['text']}, Label: {pn['label']}")

sample_text_2 = "The Eiffel Tower stands in Paris, France. John loves visiting European cities."
identified_proper_nouns_2 = find_proper_nouns(sample_text_2)

print("\nIdentified Proper Nouns (second example):")
for pn in identified_proper_nouns_2:
    print(f"  - Text: {pn['text']}, Label: {pn['label']}")
```

##Question 9: Using Genism, demonstrate how to train a simple Word2Vec model on the following dataset consisting of example sentences:
#####dataset = [
##### "Natural language processing enables computers to understand human language",
##### "Word embeddings are a type of word representation that allows words with similar
#####meaning to have similar representation",
##### "Word2Vec is a popular word embedding technique used in many NLP applications",
##### "Text preprocessing is a critical step before training word embeddings",
##### "Tokenization and normalization help clean raw text for modeling"
#####]
#####Write code that tokenizes the dataset, preprocesses it, and trains a Word2Vec model using
#####Gensim.


```python
# Install Gensim (if not already installed)
!pip install gensim nltk
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from gensim.models import Word2Vec
import string

# Download necessary NLTK data (only needs to be run once)
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

dataset = [
 "Natural language processing enables computers to understand human language",
 "Word embeddings are a type of word representation that allows words with similar meaning to have similar representation",
 "Word2Vec is a popular word embedding technique used in many NLP applications",
 "Text preprocessing is a critical step before training word embeddings",
 "Tokenization and normalization help clean raw text for modeling"
]

# Step 1 & 2: Tokenization and Preprocessing
def preprocess_text(text):
    # Lowercase the text
    text = text.lower()
    # Tokenize the text
    tokens = word_tokenize(text)
    # Remove punctuation and numbers, filter out stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [
        word for word in tokens
        if word.isalpha() and word not in stop_words
    ]
    return filtered_tokens

processed_sentences = [preprocess_text(sentence) for sentence in dataset]

print("--- Processed Sentences ---")
for sentence in processed_sentences:
    print(sentence)
print("\n")

# Step 3: Train a simple Word2Vec model
# Parameters:
#   vector_size: Dimensionality of the word vectors.
#   window: Maximum distance between the current and predicted word within a sentence.
#   min_count: Ignores all words with total frequency lower than this.
#   sg: Training algorithm: 1 for skip-gram, 0 for CBOW.
model = Word2Vec(sentences=processed_sentences, vector_size=100, window=5, min_count=1, workers=4, sg=0)

# Demonstrate some functionality of the trained model
print("--- Word2Vec Model Demonstrations ---")
# Find the most similar words to 'word'
if 'word' in model.wv.key_to_index:
    print(f"Words similar to 'word': {model.wv.most_similar('word')}")
else:
    print("'word' not in vocabulary. Try words like 'language', 'embeddings', 'nlp'.")

# You can access a word's vector directly
# Example: 'language' vector
if 'language' in model.wv.key_to_index:
    print(f"\nVector for 'language': {model.wv['language'][:5]}...") # print first 5 dimensions
else:
    print("'language' not in vocabulary.")

# Calculate similarity between two words
if 'computer' in model.wv.key_to_index and 'machine' in model.wv.key_to_index:
    print(f"\nSimilarity between 'computer' and 'machine': {model.wv.similarity('computer', 'machine')}")
else:
    print("One or both words ('computer', 'machine') not in vocabulary.")
```
